```python
f=  ('adj_Gross~ adj_Budget+o_theater+tv_movie')
error = []
df = df_for_model_1
for i in range(10):
    i
    msk =  np.random.rand(len(df))<.75
    train = df[msk]
    test =df[~msk]
    y,X = patsy.dmatrices(f, train, return_type='dataframe')
    ytest,Xtest =patsy.dmatrices(f, test, return_type='dataframe')
    model = sm.OLS(y,X).fit()
    Ypred = model.predict(Xtest)
    error.append(math.sqrt(mean_squared_error(Ypred,ytest)))
np.asanyarray(error).mean()    
```
